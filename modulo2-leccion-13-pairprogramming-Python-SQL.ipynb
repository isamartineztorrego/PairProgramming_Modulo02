```python
import pandas as pd
import mysql.connector
from mysql.connector import errorcode
```

2. Crear la conexión a la base de datos Northwind que devuelva un mensaje de error si algo no ocurre cómo debería.

```python
#Establecer conexión o printar código de error si no es posible conectarse
try:
  cnx = mysql.connector.connect(user='root', password='AlumnaAdalab', host='127.0.0.1', database='northwind')
except mysql.connector.Error as err:
  if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
    print("Something is wrong with your user name or password")
  elif err.errno == errorcode.ER_BAD_DB_ERROR:
    print("Database does not exist")
  else:
    print(err)
else:
  cnx.close()
```

3. Usando "information_schema", extraer los nombres de las columnas de la tabla "customers" de la BBDD "northwind".

```python
cnx = mysql.connector.connect(user='root', password='AlumnaAdalab', host='127.0.0.1', database='information_schema')
mycursor=cnx.cursor()
#hacemos la query
mycursor.execute('''SELECT column_name FROM COLUMNS
                 WHERE TABLE_NAME = "customers" AND table_schema = "northwind"''')
#sacamos todos los resultados de la query
myresult=mycursor.fetchall()
myresult
```

```python
#Otra forma de sacar las columnas, pero sin utilizar information_schema
cnx = mysql.connector.connect(user='root', password='AlumnaAdalab', host='127.0.0.1', database='northwind')
mycursor=cnx.cursor()
#hacemos la query
mycursor.execute('SELECT * FROM customers')
mycursor.column_names
```

4. Seleccionar todo de la tabla "customers" donde "country" sea "USA" o "Germany". Almacenar los resultados en una variable.

```python
cnx = mysql.connector.connect(user='root', password='AlumnaAdalab', host='127.0.0.1', database='northwind')
mycursor=cnx.cursor()
#hacemos la query
mycursor.execute('''SELECT * FROM customers
                 WHERE country_region in ("USA", "GERMANY")''')
myresult=mycursor.fetchall()
myresult
```
