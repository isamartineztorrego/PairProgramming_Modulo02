## Pair Programming Selenium

Descripción:

Acceso a la Página de La Casa del Libro: Utilizando Selenium, deberás automatizar el proceso de navegación hacia la página de libros de ficción en el sitio web de La Casa del Libro.

Extracción de Datos de las 5 Primeras Páginas: Una vez en la página de libros de ficción, debes programar su script para que recorra los 5 primeros libros de libros, extrayendo los siguientes datos de cada libro:

- Título del libro.

- Autor(es) del libro.

- Precio del libro.

- Editorial del libro.

- Idioma del libro.

- Número de páginas del libro.

Almacenamiento de Datos: Los datos extraídos deben almacenarse en un DataFrame.

```python
import pandas as pd

import requests

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select

from time import sleep

pd.set_option("display.max_columns", None)
```

```python
driver = webdriver.Chrome()

driver.get("https://www.casadellibro.com/")

sleep(2)

driver.maximize_window()

driver.find_element("css selector", "#onetrust-accept-btn-handler").click()

sleep(2)

driver.find_element("link text", "Ficción").click()

sleep(3)

diccionario_libros={'Título':[],
                    'Autor':[],
                    'Precio':[],
                    'Editorial':[],
                    'Idioma':[],
                    'Páginas':[]}

for i in range(1,6):
    sleep(3)
    print(f'Va por la iteración {i}')
    driver.find_element("css selector", f"#buscadorNF > div.col-md-9.col-12 > div > div.grid-view.mt-4 > div:nth-child({i}) > a > div > img").click() #se ve el listado y clicka en cada libro
    sleep (3)
    
    #algunos libros no tienen precio, así que le digo que intente coger ese selector pero si no puede, que pase.
    try:
        titulos=driver.find_element('css selector', '#app > div.v-application--wrap > main > div > div > div > div:nth-child(3) > div > div > div.col-md-5.order-lg-2.col-12.order-1 > div > h1').text
    except:
        pass
    try:
        autores=driver.find_element('css selector', '#app > div.v-application--wrap > main > div > div > div > div:nth-child(3) > div > div > div.col-md-5.order-lg-2.col-12.order-1 > div > div.text-h5.d-flex.flex-wrap.author.mb-2.justify-center.justify-sm-start > div > a > span:nth-child(1)').text
    except:
        pass
    try:
        precios=driver.find_element('css selector', '#app > div.v-application--wrap > main > div > div > div > div:nth-child(3) > div > div > div.border-left.col-md-4.col-12.order-3 > div > div:nth-child(3) > div.col.col-8 > strong').text
    except:
        pass
    try:
        editoriales=driver.find_element('css selector', '#app > div.v-application--wrap > main > div > div > div > div:nth-child(7) > div > div > div.caracteristicas.mt-4.pt-4 > div:nth-child(3) > div').text
    except:
        pass
    try:
        idiomas=driver.find_element('css selector', '#app > div.v-application--wrap > main > div > div > div > div:nth-child(7) > div > div > div.caracteristicas.mt-4.pt-4 > div:nth-child(4) > div').text
    except:
        pass
    try:
        paginas=driver.find_element('css selector', '#app > div.v-application--wrap > main > div > div > div > div:nth-child(7) > div > div > div.caracteristicas.mt-4.pt-4 > div:nth-child(1) > div').text
    except:
        pass

    diccionario_libros['Título'].append(titulos)
    diccionario_libros['Autor'].append(autores)
    diccionario_libros['Precio'].append(precios)
    diccionario_libros['Editorial'].append(editoriales)
    diccionario_libros['Idioma'].append(idiomas)
    diccionario_libros['Páginas'].append(paginas)

    sleep(3)
    driver.back()
```
